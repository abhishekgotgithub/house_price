```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_csv('C:/Users/Admin/PycharmProjects/pythonProject3/house_price/data/HousingData.csv')
corrmatrix=df.corr()
print(corrmatrix)

```

```python
# creating a heatmap for corelational matrix
plt.figure(figsize=(15,15))
sns.heatmap(corrmatrix,
           xticklabels=df.columns,
           yticklabels=df.columns,
           cmap='RdBu_r',
           annot=True,
           linewidth=0.5)

```

```python
sns.pairplot(df) # creating pair plots

# histogram and density plot
for key,values in df.iteritems():
    sns.distplot(df[key],kde=False)
    plt.show()

```

```python
# box plot
for key,values in df.iteritems():
    sns.boxplot(df[key])
    plt.show()
```
